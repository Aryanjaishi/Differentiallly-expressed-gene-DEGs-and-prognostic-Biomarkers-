```python
#A ball is thrown vertically up in the air from a height $h_{0}$ above the ground at an initial velocity $v_{0}$. Its subsequent height $h$ and velocity $v$ are given by the equations $h=h_{0}+v_{0}t-\frac{1}{2}gt^{2}$ and $v=v_{0}-gt$, where $g=9.8$ is the acceleration due to gravity in $m/s^{2}$. Write a script that finds the height $h$ and velocity $v$ at a time $t$ after the ball is thrown. Start the code by setting $h_{0}=1.2$ (meters) and $v_{0}=5.4 (m/s)$ and have your code print out the values of height and velocity. Then use the script to find the height and velocity after 0.5 seconds. Then modify your script to find them after 2.0 seconds.
# 1. A ball is thrown vertically up from height h0 with initial velocity v0.
# We need to find height (h) and velocity (v) at time (t).

# Given values
g = 9.8          # Gravity in m/s^2 
h0 = 1.2         # Starting height in meters
v0 = 5.4         # Starting velocity in m/s 

# --- Calculation for t = 0.5 seconds ---
t1 = 0.5
h1 = h0 + (v0 * t1) - (0.5 * g * t1**2) # Height formula
v1 = v0 - (g * t1)                      # Velocity formula 

print("At 0.5 seconds:")
print("Height:", h1, "meters")
print("Velocity:", v1, "m/s")

# Calculation for t = 2.0 seconds
t2 = 2.0
h2 = h0 + (v0 * t2) - (0.5 * g * t2**2)
v2 = v0 - (g * t2)

print("\nAt 2.0 seconds:")
print("Height:", h2, "meters")
print("Velocity:", v2, "m/s")
```

```python
# 4. The position of a ball at time t dropped with zero initial velocity from a height h0 is given by y = h0 - 0.5gt^2, where g = 9.8 m/s^2. Suppose h0 = 10 m. Find the sequence of times when the ball passes each half meter. Solving the above equation for t, show that t = sqrt(2(h0 - y)) / g. Using this equation and the array you created, find the sequence of times when the ball passes each half meter. Recalling that the average velocity over an interval delta t is defined as v_avg = delta y / delta t, find the average velocity for each time interval in the problem using NumPy arrays. [cite: 24, 25, 26, 30, 31, 32, 33, 34]
import numpy as np

# Constants
h0 = 10.0
g = 9.8

# Create an array of y values from 10 to 0 in increments of -0.5 [cite: 26, 29]
y = np.arange(10, -0.5, -0.5)

# Calculate the sequence of times 
# Using the derived formula t = sqrt(2 * (h0 - y) / g)
t = np.sqrt(2 * (h0 - y) / g)

print("Times when the ball passes each half meter:")
print(t)

# Calculate average velocity for each interval 
# Average velocity = delta_y / delta_t
delta_y = np.diff(y)
delta_t = np.diff(t)
avg_velocity = delta_y / delta_t

print("\nAverage velocity for each time interval:")
print(avg_velocity)
```

```python
#Write a program to tell the nature of the roots and values of the roots of a quadratic 
#equation 𝑎𝑥^2 + 𝑏𝑥 + 𝑐 = 0, 𝑎 ≠ 0. 
import cmath

# Input coefficients from the user
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Calculate the discriminant
d = (b**2) - (4*a*c)

# Determine the nature of the roots
if d > 0:
    root1 = (-b + d**0.5) / (2*a)
    root2 = (-b - d**0.5) / (2*a)
    print("Nature: Two distinct real roots.")
    print(f"Roots: {root1} and {root2}")
elif d == 0:
    root = -b / (2*a)
    print("Nature: One real root (repeated).")
    print(f"Root: {root}")
else:
    root1 = (-b + cmath.sqrt(d)) / (2*a)
    root2 = (-b - cmath.sqrt(d)) / (2*a)
    print("Nature: Two complex roots.")
    print(f"Roots: {root1} and {root2}")
```

```python
#Write a program to calculate the factorial of a positive integer input by the user.
#(a) Write the factorial function using a Python while loop.
#(b) Write the factorial function using a Python for loop.
#Check your programs to make sure they work for 0, 1, 2, 3, 4, 5.
# (a) Factorial using a while loop
def factorial_while(n):
    result = 1
    i = 1
    while i <= n:
        result *= i
        i += 1
    return result

# (b) Factorial using a for loop
def factorial_for(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Testing the programs for 0, 1, 2, 3, 4, 5
test_values = [0, 1, 2, 3, 4, 5]

print("Testing Factorial Functions:")
for val in test_values:
    print(f"n={val}: While loop result = {factorial_while(val)}, For loop result = {factorial_for(val)}")
```

```python
#Write a program to calculate the sum of natural n natural numbers 1 + 2 + ⋯ + 𝑛. 
#Calculate the sum when 𝑛 = 10.
# Program to calculate the sum of n natural numbers 1 + 2 + ... + n

def calculate_sum(n):
    # Sum using built-in range and sum function
    return sum(range(1, n + 1))

n = 10
total = calculate_sum(n)

print(f"The sum of the first {n} natural numbers is: {total}")
```

```python
#Write a program to calculate the sum of squares of natural n natural numbers 1^2 + 2^2 +⋯ + 𝑛^2. Calculate the sum when 𝑛 = 5.
# Program to calculate the sum of squares: 1^2 + 2^2 + ... + n^2
def sum_of_squares(n):
    total = 0
    for i in range(1, n + 1):
        total += i**2
    return total

# Calculate the sum when n = 5
n = 5
result = sum_of_squares(n)

print(f"The sum of squares of the first {n} natural numbers is: {result}")
```

```python
# 11. Plot the functions y=sin x and y=cos x for -2pi <= x <= 2pi on the same plot.
# Plot y=sin x in the color red and y=cos x in the color blue and include a legend to label the two curves.
# Place the legend within the plot, but such that it does not cover either of the sine or cosine traces.

import matplotlib.pyplot as plt
import numpy as np

# Create x values from -2pi to 2pi
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)

# Calculate y values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plotting
plt.plot(x, y_sin, color='red', label='y = sin(x)')
plt.plot(x, y_cos, color='blue', label='y = cos(x)')

# Add legend (placed in a location that avoids overlapping traces)
plt.legend(loc='upper right')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of sin(x) and cos(x)')

# Display the plot
plt.show()
```

```python
# 12. Plot the space curve x=2*theta*cos(theta), y=2*theta*sin(theta), z=1+2*theta for 0 <= theta <= 20*pi 

import matplotlib.pyplot as plt
import numpy as np

# Generate theta values from 0 to 20*pi [cite: 49]
theta = np.linspace(0, 20 * np.pi, 1000)

# Define the equations for the space curve [cite: 49]
x = 2 * theta * np.cos(theta)
y = 2 * theta * np.sin(theta)
z = 1 + 2 * theta

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the curve
ax.plot(x, y, z)

# Label the axes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Space Curve')

# Display the plot
plt.show()
```

```python
# 13. Write the program for (i) surface plot (ii) contour plot on xy-plane of the function 
# f(x, y) = (1 - x/2 + x^5 + y^5) / e^(-x^2 - y^2) 
# on the domain -3 <= x <= 3 and -3 <= y <= 3

import matplotlib.pyplot as plt
import numpy as np

# Create the domain for x and y
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Define the function f(x, y)
Z = (1 - (X / 2) + X**5 + Y**5) / np.exp(-X**2 - Y**2)

# (i) Surface Plot
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Surface Plot of f(x,y)')
plt.show()

# (ii) Contour Plot
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
contour = ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.set_title('Contour Plot of f(x,y)')
fig2.colorbar(contour)
plt.show()
```
