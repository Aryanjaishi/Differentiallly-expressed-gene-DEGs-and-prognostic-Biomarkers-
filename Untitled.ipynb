```python
from numpy import linspace, zeros, exp
import matplotlib.pyplot as plt

def f(x, y):
    return 3 - 2*x - y/2   

n = 5
x = linspace(0, 2, n+1)   
h = x[1] - x[0]           
y = zeros(n+1)
y[0] = 1                  


for i in range(n):
    y[i+1] = y[i] + h * f(x[i], y[i])


plt.plot(x, y, "ro--", label="Euler Approximation")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler's Method Approximation")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ODE dy/dx = 3 - 2x - y/2
def f(x, y):
    return 3 - 2*x - y/2

# Setup
n = 5
x = np.linspace(0, 2, n+1)   # x values
h = x[1] - x[0]              # step size
y = np.zeros(n+1)
y[0] = 1                     # initial condition

# Improved Euler (Heun's method)
for i in range(n):
    k1 = h * f(x[i], y[i])
    k2 = h * f(x[i] + h, y[i] + k1)
    y[i+1] = y[i] + 0.5 * (k1 + k2)

# Plot
plt.plot(x, y, "ro--", label="Improved Euler Approximation")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Improved Euler ")
plt.legend()
plt.grid(True)
plt.show()
```
