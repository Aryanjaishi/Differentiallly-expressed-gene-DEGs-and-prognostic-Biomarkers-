```python
import pandas as pd
import numpy as np

# Load the cleaned untreated (Placebo) data from Step 10
df = pd.read_csv(r"C:\jupyter demo\cleaned_untreated_tumor_growth.csv")

# Quick check
df.head(10)
```

```python
from scipy.optimize import curve_fit

# This is the Gompertz formula written as a Python function
def gompertz(t, K, alpha, N0):
    return K * np.exp(np.log(N0 / K) * np.exp(-alpha * t))

# Let's pick just ONE mouse to test on first
one_mouse_id = df["mouse_id"].unique()[0]   # grabs the first mouse's ID
one_mouse = df[df["mouse_id"] == one_mouse_id].sort_values("day")

print("Testing on mouse:", one_mouse_id)
print(one_mouse[["day", "volume"]])

# Get this mouse's day and volume values, and its starting size N0
t_data = one_mouse["day"].values
y_data = one_mouse["volume"].values
N0 = y_data[0]   # volume at day 0

print("\nN0 (starting volume) for this mouse:", N0)
```

```python
# curve_fit needs a function where K and alpha are the only unknowns
# so we "wrap" our gompertz function to lock in this mouse's N0
def gompertz_wrapped(t, K, alpha):
    return gompertz(t, K, alpha, N0)

# Starting guesses for K and alpha (curve_fit needs a starting point to search from)
initial_guess = [y_data.max() * 3, 0.05]   # guess K = 3x the biggest volume seen, alpha = 0.05

# Run the fit
params, _ = curve_fit(gompertz_wrapped, t_data, y_data, p0=initial_guess, maxfev=5000)

K_fit, alpha_fit = params
print("Fitted K (carrying capacity):", K_fit)
print("Fitted alpha (growth rate):", alpha_fit)
```

```python
import matplotlib.pyplot as plt

# Generate a smooth range of days to draw the fitted curve
t_smooth = np.linspace(0, t_data.max(), 100)   # 100 evenly spaced days from 0 to the last day
volume_predicted = gompertz_wrapped(t_smooth, K_fit, alpha_fit)

# Plot: real data points vs. the fitted curve
plt.figure(figsize=(7, 5))
plt.scatter(t_data, y_data, color="black", label="Actual data (real measurements)")
plt.plot(t_smooth, volume_predicted, color="blue", label="Gompertz fitted curve")
plt.xlabel("Day")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Gompertz Fit — Mouse {one_mouse_id}")
plt.legend()
plt.show()
```

```python
# This will store the results for every mouse
results = []

# Loop through each mouse, one at a time
for mouse_id in df["mouse_id"].unique():

    # Get just this mouse's data, sorted by day
    mouse_data = df[df["mouse_id"] == mouse_id].sort_values("day")
    t_data = mouse_data["day"].values
    y_data = mouse_data["volume"].values
    N0 = y_data[0]

    # Wrap the formula for this specific mouse's N0 (same idea as Step 3)
    def gompertz_wrapped(t, K, alpha, N0=N0):
        return gompertz(t, K, alpha, N0)

    # Try to fit; if it fails for some reason, skip it instead of crashing
    try:
        initial_guess = [y_data.max() * 3, 0.05]
        params, _ = curve_fit(gompertz_wrapped, t_data, y_data, p0=initial_guess, maxfev=5000)
        K_fit, alpha_fit = params

        results.append({
            "mouse_id": mouse_id,
            "N0": N0,
            "K": K_fit,
            "alpha": alpha_fit,
            "success": True
        })
    except Exception as e:
        results.append({
            "mouse_id": mouse_id,
            "N0": N0,
            "K": None,
            "alpha": None,
            "success": False
        })

# Turn the results into a table
results_df = pd.DataFrame(results)
print(results_df)

# How many mice fitted successfully?
print("\nSuccessful fits:", results_df["success"].sum(), "out of", len(results_df))
```

```python
# Save every mouse's individual fit to a CSV file
results_df.to_csv(r"C:\jupyter demo\fitted_gompertz_parameters.csv", index=False)
print("Saved: fitted_gompertz_parameters.csv")

# Keep only the mice that fitted successfully
good_fits = results_df[results_df["success"] == True]

# Calculate the average and spread (standard deviation) of K and alpha
alpha_mean = good_fits["alpha"].mean()
alpha_std = good_fits["alpha"].std()
K_mean = good_fits["K"].mean()
K_std = good_fits["K"].std()

print("\n--- Population-level Gompertz parameters ---")
print(f"alpha: mean = {alpha_mean:.5f}, std dev = {alpha_std:.5f}")
print(f"K:     mean = {K_mean:.2f}, std dev = {K_std:.2f}")

# Save this summary too
summary_df = pd.DataFrame({
    "parameter": ["alpha", "K"],
    "mean": [alpha_mean, K_mean],
    "std_dev": [alpha_std, K_std]
})
summary_df.to_csv(r"C:\jupyter demo\population_alpha_K_summary.csv", index=False)
print("\nSaved: population_alpha_K_summary.csv")
```

```python
from scipy.integrate import odeint

# This is the drug-response equation, written as Python code
# It describes HOW FAST the tumor is growing/shrinking at any moment
def gompertz_with_drug(N, t, k_eff, alpha, K):
    growth = alpha * N * np.log(K / N)   # normal Gompertz growth (same as before)
    kill = k_eff * N                      # drug effect: kills a portion of the tumor
    return growth - kill

# Pick ONE treated mouse to test on first (just like we did for untreated)
one_treated_id = treated_df["mouse_id"].unique()[0]
one_treated = treated_df[treated_df["mouse_id"] == one_treated_id].sort_values("day")

t_data = one_treated["day"].values
y_data = one_treated["volume"].values
N0 = y_data[0]
regimen_name = one_treated["regimen"].iloc[0]

print("Testing on mouse:", one_treated_id, "| Drug:", regimen_name)
print(one_treated[["day", "volume"]])

# Try a GUESSED value for k_eff (just to see what happens - not fitted yet)
k_eff_guess = 0.05

# Simulate the tumor's volume day by day, using this guess
simulated_volume = odeint(gompertz_with_drug, N0, t_data, args=(k_eff_guess, alpha_mean, K_mean))

print("\nSimulated volumes with guessed k_eff =", k_eff_guess)
print(simulated_volume.flatten())
print("\nActual real volumes:")
print(y_data)
```

```python
from scipy.optimize import minimize_scalar

# This function checks: "if I use this k_eff, how far off is the simulation from real data?"
def error_for_this_k_eff(k_eff):
    simulated = odeint(gompertz_with_drug, N0, t_data, args=(k_eff, alpha_mean, K_mean))
    simulated = simulated.flatten()
    error = np.sum((y_data - simulated) ** 2)   # total squared difference
    return error

# Search for the k_eff value that gives the SMALLEST error
# We tell it to search only between 0 and 1 (a reasonable range for this kind of rate)
result = minimize_scalar(error_for_this_k_eff, bounds=(0, 1), method="bounded")

best_k_eff = result.x
print("Best-fit k_eff for this mouse:", best_k_eff)

# Let's see how well this best k_eff matches the real data
simulated_best = odeint(gompertz_with_drug, N0, t_data, args=(best_k_eff, alpha_mean, K_mean))
print("\nSimulated volumes (using best k_eff):")
print(simulated_best.flatten())
print("\nActual real volumes:")
print(y_data)
```

```python
# This will store the results for every treated mouse
drug_results = []

# Loop through each treated mouse, one at a time
for mouse_id in treated_df["mouse_id"].unique():

    # Get this mouse's data, sorted by day
    mouse_data = treated_df[treated_df["mouse_id"] == mouse_id].sort_values("day")
    t_data = mouse_data["day"].values
    y_data = mouse_data["volume"].values
    N0 = y_data[0]
    regimen_name = mouse_data["regimen"].iloc[0]

    # Same error-checking function as Step 3, just using this mouse's own data
    def error_for_this_k_eff(k_eff, t_data=t_data, y_data=y_data, N0=N0):
        simulated = odeint(gompertz_with_drug, N0, t_data, args=(k_eff, alpha_mean, K_mean))
        simulated = simulated.flatten()
        return np.sum((y_data - simulated) ** 2)

    # Try to find the best k_eff; skip this mouse if something goes wrong
    try:
        result = minimize_scalar(error_for_this_k_eff, bounds=(0, 1), method="bounded")
        drug_results.append({
            "mouse_id": mouse_id,
            "regimen": regimen_name,
            "N0": N0,
            "k_eff": result.x,
            "success": True
        })
    except Exception as e:
        drug_results.append({
            "mouse_id": mouse_id,
            "regimen": regimen_name,
            "N0": N0,
            "k_eff": None,
            "success": False
        })

# Turn results into a table
drug_results_df = pd.DataFrame(drug_results)
print(drug_results_df.head(10))
print("\nSuccessful fits:", drug_results_df["success"].sum(), "out of", len(drug_results_df))
```

```python
import numpy as np

# ---- Parameters (you can adjust these to match your fitted values later) ----
alpha = 0.05     # intrinsic growth rate
K     = 1500.0   # carrying capacity (maximum tumor size)
N0    = 45.0     # starting tumor volume
t0, tf = 0.0, 60.0   # simulate from day 0 to day 60
dt = 0.1              # step size (smaller = more accurate, but slower)

# ---- Standard Form (Eq. 1): dN/dt = alpha * N * ln(K/N) ----
def gompertz_standard(t, N):
    N = max(N, 1e-8)   # safety: stop N from ever being 0 or negative (log(0) breaks)
    return alpha * N * np.log(K / N)
```

```python
# ---- A reusable RK4 solver ----
# This works for ANY equation of the form dN/dt = f(t, N), not just Gompertz
def solve_rk4(f, t0, tf, N0, dt):
    n_steps = int(np.round((tf - t0) / dt))       # how many small steps we need
    t_vals = np.linspace(t0, tf, n_steps + 1)      # list of every time point we'll calculate
    N_vals = np.zeros(n_steps + 1)                 # empty list to store the volume at each time point
    N_vals[0] = N0                                 # the very first value is just our starting size

    for i in range(n_steps):
        t_i, N_i = t_vals[i], N_vals[i]
        k1 = f(t_i, N_i)
        k2 = f(t_i + dt/2, N_i + dt/2 * k1)
        k3 = f(t_i + dt/2, N_i + dt/2 * k2)
        k4 = f(t_i + dt,   N_i + dt * k3)
        N_vals[i+1] = N_i + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)

    return t_vals, N_vals

# ---- Use it to solve the Standard Form (Eq. 1) ----
t_standard, N_standard = solve_rk4(gompertz_standard, t0, tf, N0, dt)

print("Tumor volume at day 60:", N_standard[-1])
```

```python
# ---- Retardation Form (Eq. 2): dN/dt = N * [a - beta*ln(N)] ----
#
# IMPORTANT: this equation uses "alpha" to mean something DIFFERENT from
# the alpha in the Standard Form (Eq. 1). We worked this out earlier:
#
#   Expanding Eq. 1:  alpha*N*ln(K/N) = N*(alpha*ln(K) - alpha*ln(N))
#
#   So matching to Eq. 2's shape N*(a - beta*ln(N)) gives:
#       a    = alpha * ln(K)     <-- this is what Eq. 2 calls "alpha"
#       beta = a / ln(K)         <-- this works out to equal Eq. 1's alpha

a_eq2 = alpha * np.log(K)     # Eq. 2's constant term
beta  = a_eq2 / np.log(K)     # Eq. 2's ln(N) coefficient (per the proposal's own formula)

print("Eq.1 alpha:", alpha)
print("Eq.2 'a' term (alpha*ln(K)):", a_eq2)
print("Eq.2 beta:", beta, " <- should equal Eq.1's alpha")

def gompertz_retardation(t, N):
    N = max(N, 1e-8)
    return N * (a_eq2 - beta * np.log(N))

# ---- Solve it using the SAME RK4 function from Step 2 ----
t_retard, N_retard = solve_rk4(gompertz_retardation, t0, tf, N0, dt)

print("\nTumor volume at day 60 (Retardation Form):", N_retard[-1])
print("Tumor volume at day 60 (Standard Form, from Step 2):", N_standard[-1])
print("Difference:", abs(N_retard[-1] - N_standard[-1]))
```

```python
import numpy as np

# ---- Step 1: Parameters + Standard Form ----
alpha = 0.05
K     = 1500.0
N0    = 45.0
t0, tf = 0.0, 60.0
dt = 0.1

def gompertz_standard(t, N):
    N = max(N, 1e-8)
    return alpha * N * np.log(K / N)

# ---- Step 2: RK4 solver, solving Standard Form ----
def solve_rk4(f, t0, tf, N0, dt):
    n_steps = int(np.round((tf - t0) / dt))
    t_vals = np.linspace(t0, tf, n_steps + 1)
    N_vals = np.zeros(n_steps + 1)
    N_vals[0] = N0
    for i in range(n_steps):
        t_i, N_i = t_vals[i], N_vals[i]
        k1 = f(t_i, N_i)
        k2 = f(t_i + dt/2, N_i + dt/2 * k1)
        k3 = f(t_i + dt/2, N_i + dt/2 * k2)
        k4 = f(t_i + dt,   N_i + dt * k3)
        N_vals[i+1] = N_i + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)
    return t_vals, N_vals

t_standard, N_standard = solve_rk4(gompertz_standard, t0, tf, N0, dt)
print("Standard Form - tumor volume at day 60:", N_standard[-1])

# ---- Step 3: Retardation Form ----
a_eq2 = alpha * np.log(K)
beta  = a_eq2 / np.log(K)

def gompertz_retardation(t, N):
    N = max(N, 1e-8)
    return N * (a_eq2 - beta * np.log(N))

t_retard, N_retard = solve_rk4(gompertz_retardation, t0, tf, N0, dt)
print("Retardation Form - tumor volume at day 60:", N_retard[-1])
print("Difference:", abs(N_retard[-1] - N_standard[-1]))
```

```python
# ---- Closed-Form Analytical Solution (Eq. 3) ----
# N(t) = K * exp( ln(N0/K) * exp(-alpha*t) )
#
# This is the EXACT solution to Eq. 1, solved by hand using algebra
# (no simulation/stepping needed - just plug in t and get the answer directly)

def gompertz_analytical(t, K, alpha, N0):
    return K * np.exp(np.log(N0 / K) * np.exp(-alpha * t))

# Calculate it at the same time points as our RK4 solutions, so we can compare
N_analytical = gompertz_analytical(t_standard, K, alpha, N0)

# ---- Compare all three: Standard Form (RK4), Retardation Form (RK4), Analytical ----
print("At day 60:")
print("  Standard Form (RK4):     ", N_standard[-1])
print("  Retardation Form (RK4):  ", N_retard[-1])
print("  Analytical (exact):      ", N_analytical[-1])

max_error_standard = np.max(np.abs(N_standard - N_analytical))
max_error_retard = np.max(np.abs(N_retard - N_analytical))
print("\nMax error, Standard Form vs Analytical:", max_error_standard)
print("Max error, Retardation Form vs Analytical:", max_error_retard)
```

```python
# ---- Drug Response Extension (Eq. 4): dN/dt = alpha*N*ln(K/N) - delta*C(t)*N ----
#
# This equation needs a drug concentration C(t) to work. Before building the
# full coupled system (Step 6-7), let's first understand Eq. 4 on its own by
# using a SIMPLE placeholder for C(t): just a constant drug level, always present.
# This isn't realistic yet (real drug levels rise and fall) - it's just to see
# how the drug term behaves before we add the more complex pharmacokinetics.

delta = 0.0005   # drug-kill rate constant (how strong the drug is per unit concentration)
C_constant = 20.0  # a constant, simplified drug concentration (placeholder for now)

def gompertz_with_drug_simple(t, N):
    N = max(N, 1e-8)
    growth = alpha * N * np.log(K / N)     # same normal growth as Eq. 1
    kill = delta * C_constant * N           # NEW: drug kills a portion of the tumor
    return growth - kill

# Solve it using the SAME RK4 function from before - no changes needed there
t_drug_simple, N_drug_simple = solve_rk4(gompertz_with_drug_simple, t0, tf, N0, dt)

print("Tumor volume at day 60:")
print("  Untreated (Standard Form):", N_standard[-1])
print("  Treated (constant drug level):", N_drug_simple[-1])
```

```python
# ---- Drug Pharmaco-kinetics (Eq. 5): dC/dt = -lambda * C(t) + u(t) ----
#
# This describes how drug concentration actually behaves in the body:
# it goes UP when a dose is given (u(t)), and goes DOWN steadily over
# time as the body eliminates it (-lambda * C).

lam = 0.3   # drug elimination rate (how fast the body clears the drug)
            # named "lam" instead of "lambda" because lambda is a reserved word in Python

# ---- The dosing function u(t): when and how much drug is given ----
dose_amount   = 5.0    # how much drug is delivered per dose
dose_interval = 7.0    # days between doses (e.g. once a week)
pulse_width   = 0.5    # how many days each dose "pulse" lasts (short = closer to a real injection)
first_dose_day = 5.0   # treatment starts on day 5

def u(t):
    if t < first_dose_day:
        return 0.0   # no drug given yet
    time_since_first = t - first_dose_day
    time_in_cycle = time_since_first % dose_interval   # where are we in the weekly cycle?
    if time_in_cycle < pulse_width:
        return dose_amount / pulse_width   # currently delivering a dose
    return 0.0   # between doses - no new drug being added

# ---- Eq. 5 written as code ----
def drug_pk(t, C):
    C = max(C, 0.0)   # drug concentration can't be negative
    return -lam * C + u(t)

# ---- Solve it on its own first, just to see what the concentration looks like ----
C0 = 0.0   # no drug in the body at the start
t_pk, C_pk = solve_rk4(drug_pk, t0, tf, C0, dt)

# Plot it
plt.figure(figsize=(9, 4))
plt.plot(t_pk, C_pk, color="green")
plt.axvline(first_dose_day, color="red", linestyle=":", label="Treatment starts")
plt.xlabel("Day")
plt.ylabel("Drug Concentration C(t)")
plt.title("Drug Concentration Over Time ")
plt.legend()
plt.show()
```

```python
# ---- Coupled System: tumor (Eq. 4) AND drug (Eq. 5) solved TOGETHER ----
#
# Until now, we solved N and C separately. But really, they affect each
# other: C(t) determines how hard the drug hits the tumor in Eq. 4,
# and both are changing at the same time. So now we track BOTH together
# as a pair, at every single time step.

def coupled_system(t, y):
    N, C = y   # unpack: y is now a PAIR of numbers [tumor volume, drug concentration]
    N = max(N, 1e-8)
    C = max(C, 0.0)

    dN = alpha * N * np.log(K / N) - delta * C * N   # Eq. 4 - now uses the REAL changing C(t)
    dC = -lam * C + u(t)                              # Eq. 5 - same as Step 6

    return np.array([dN, dC])   # return BOTH rates of change together

# ---- RK4 solver, adapted slightly to handle a PAIR of values instead of one ----
def solve_rk4_system(f, t0, tf, y0, dt):
    n_steps = int(np.round((tf - t0) / dt))
    t_vals = np.linspace(t0, tf, n_steps + 1)
    y_vals = np.zeros((n_steps + 1, len(y0)))   # a table: one row per time step, one column per variable (N and C)
    y_vals[0] = y0

    for i in range(n_steps):
        t_i, y_i = t_vals[i], y_vals[i]
        k1 = f(t_i, y_i)
        k2 = f(t_i + dt/2, y_i + dt/2 * k1)
        k3 = f(t_i + dt/2, y_i + dt/2 * k2)
        k4 = f(t_i + dt,   y_i + dt * k3)
        y_vals[i+1] = y_i + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)

    return t_vals, y_vals

# ---- Solve the full coupled system ----
y0 = [N0, 0.0]   # starting point: tumor = N0, drug concentration = 0
t_coupled, y_coupled = solve_rk4_system(coupled_system, t0, tf, y0, dt)

N_coupled = y_coupled[:, 0]   # first column = tumor volume over time
C_coupled = y_coupled[:, 1]   # second column = drug concentration over time

print("Final tumor volume at day 60:")
print("  Untreated:", N_standard[-1])
print("  Treated (constant drug, Step 5):", N_drug_simple[-1])
print("  Treated (REAL changing drug, Step 7):", N_coupled[-1])
```

```python
from scipy.optimize import curve_fit

# This is the Gompertz formula written as a Python function
def gompertz(t, K, alpha, N0):
    return K * np.exp(np.log(N0 / K) * np.exp(-alpha * t))

# Let's pick just ONE mouse to test on first
one_mouse_id = df["mouse_id"].unique()[0]   # grabs the first mouse's ID
one_mouse = df[df["mouse_id"] == one_mouse_id].sort_values("day")

print("Testing on mouse:", one_mouse_id)
print(one_mouse[["day", "volume"]])

# Get this mouse's day and volume values, and its starting size N0
t_data = one_mouse["day"].values
y_data = one_mouse["volume"].values
N0 = y_data[0]   # volume at day 0

print("\nN0 (starting volume) for this mouse:", N0)
```

```python
# curve_fit needs a function where K and alpha are the only unknowns
# so we "wrap" our gompertz function to lock in this mouse's N0
def gompertz_wrapped(t, K, alpha):
    return gompertz(t, K, alpha, N0)

# Starting guesses for K and alpha (curve_fit needs a starting point to search from)
initial_guess = [y_data.max() * 3, 0.05]   # guess K = 3x the biggest volume seen, alpha = 0.05

# Run the fit
params, _ = curve_fit(gompertz_wrapped, t_data, y_data, p0=initial_guess, maxfev=5000)

K_fit, alpha_fit = params
print("Fitted K (carrying capacity):", K_fit)
print("Fitted alpha (growth rate):", alpha_fit)
```

```python
import matplotlib.pyplot as plt

# Generate a smooth range of days to draw the fitted curve
t_smooth = np.linspace(0, t_data.max(), 100)   # 100 evenly spaced days from 0 to the last day
volume_predicted = gompertz_wrapped(t_smooth, K_fit, alpha_fit)

# Plot: real data points vs. the fitted curve
plt.figure(figsize=(7, 5))
plt.scatter(t_data, y_data, color="black", label="Actual data (real measurements)")
plt.plot(t_smooth, volume_predicted, color="blue", label="Gompertz fitted curve")
plt.xlabel("Day")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Gompertz Fit — Mouse {one_mouse_id}")
plt.legend()
plt.show()
```

```python
# This will store the results for every mouse
results = []

# Loop through each mouse, one at a time
for mouse_id in df["mouse_id"].unique():

    # Get just this mouse's data, sorted by day
    mouse_data = df[df["mouse_id"] == mouse_id].sort_values("day")
    t_data = mouse_data["day"].values
    y_data = mouse_data["volume"].values
    N0 = y_data[0]

    # Wrap the formula for this specific mouse's N0 (same idea as Step 3)
    def gompertz_wrapped(t, K, alpha, N0=N0):
        return gompertz(t, K, alpha, N0)

    # Try to fit; if it fails for some reason, skip it instead of crashing
    try:
        initial_guess = [y_data.max() * 3, 0.05]
        params, _ = curve_fit(gompertz_wrapped, t_data, y_data, p0=initial_guess, maxfev=5000)
        K_fit, alpha_fit = params

        results.append({
            "mouse_id": mouse_id,
            "N0": N0,
            "K": K_fit,
            "alpha": alpha_fit,
            "success": True
        })
    except Exception as e:
        results.append({
            "mouse_id": mouse_id,
            "N0": N0,
            "K": None,
            "alpha": None,
            "success": False
        })

# Turn the results into a table
results_df = pd.DataFrame(results)
print(results_df)

# How many mice fitted successfully?
print("\nSuccessful fits:", results_df["success"].sum(), "out of", len(results_df))
```

```python
# ---- A reusable RK4 solver ----
# This works for ANY equation of the form dN/dt = f(t, N), not just Gompertz
def solve_rk4(f, t0, tf, N0, dt):
    n_steps = int(np.round((tf - t0) / dt))       # how many small steps we need
    t_vals = np.linspace(t0, tf, n_steps + 1)      # list of every time point we'll calculate
    N_vals = np.zeros(n_steps + 1)                 # empty list to store the volume at each time point
    N_vals[0] = N0                                 # the very first value is just our starting size

    for i in range(n_steps):
        t_i, N_i = t_vals[i], N_vals[i]
        k1 = f(t_i, N_i)
        k2 = f(t_i + dt/2, N_i + dt/2 * k1)
        k3 = f(t_i + dt/2, N_i + dt/2 * k2)
        k4 = f(t_i + dt,   N_i + dt * k3)
        N_vals[i+1] = N_i + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)

    return t_vals, N_vals

# ---- Use it to solve the Standard Form (Eq. 1) ----
t_standard, N_standard = solve_rk4(gompertz_standard, t0, tf, N0, dt)

print("Tumor volume at day 60:", N_standard[-1])
```

```python
import numpy as np

# ---- Step 1: Parameters + Standard Form ----
alpha = 0.05
K     = 1500.0
N0    = 45.0
t0, tf = 0.0, 60.0
dt = 0.1

def gompertz_standard(t, N):
    N = max(N, 1e-8)
    return alpha * N * np.log(K / N)

# ---- Step 2: RK4 solver, solving Standard Form ----
def solve_rk4(f, t0, tf, N0, dt):
    n_steps = int(np.round((tf - t0) / dt))
    t_vals = np.linspace(t0, tf, n_steps + 1)
    N_vals = np.zeros(n_steps + 1)
    N_vals[0] = N0
    for i in range(n_steps):
        t_i, N_i = t_vals[i], N_vals[i]
        k1 = f(t_i, N_i)
        k2 = f(t_i + dt/2, N_i + dt/2 * k1)
        k3 = f(t_i + dt/2, N_i + dt/2 * k2)
        k4 = f(t_i + dt,   N_i + dt * k3)
        N_vals[i+1] = N_i + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)
    return t_vals, N_vals

t_standard, N_standard = solve_rk4(gompertz_standard, t0, tf, N0, dt)
print("Standard Form - tumor volume at day 60:", N_standard[-1])

# ---- Step 3: Retardation Form ----
a_eq2 = alpha * np.log(K)
beta  = a_eq2 / np.log(K)

def gompertz_retardation(t, N):
    N = max(N, 1e-8)
    return N * (a_eq2 - beta * np.log(N))

t_retard, N_retard = solve_rk4(gompertz_retardation, t0, tf, N0, dt)
print("Retardation Form - tumor volume at day 60:", N_retard[-1])
print("Difference:", abs(N_retard[-1] - N_standard[-1]))
```

```python
# ---- Closed-Form Analytical Solution (Eq. 3) ----
# N(t) = K * exp( ln(N0/K) * exp(-alpha*t) )
#
# This is the EXACT solution to Eq. 1, solved by hand using algebra
# (no simulation/stepping needed - just plug in t and get the answer directly)

def gompertz_analytical(t, K, alpha, N0):
    return K * np.exp(np.log(N0 / K) * np.exp(-alpha * t))

# Calculate it at the same time points as our RK4 solutions, so we can compare
N_analytical = gompertz_analytical(t_standard, K, alpha, N0)

# ---- Compare all three: Standard Form (RK4), Retardation Form (RK4), Analytical ----
print("At day 60:")
print("  Standard Form (RK4):     ", N_standard[-1])
print("  Retardation Form (RK4):  ", N_retard[-1])
print("  Analytical (exact):      ", N_analytical[-1])

max_error_standard = np.max(np.abs(N_standard - N_analytical))
max_error_retard = np.max(np.abs(N_retard - N_analytical))
print("\nMax error, Standard Form vs Analytical:", max_error_standard)
print("Max error, Retardation Form vs Analytical:", max_error_retard)
```

```python
# ---- Drug Response Extension (Eq. 4): dN/dt = alpha*N*ln(K/N) - delta*C(t)*N ----
#
# This equation needs a drug concentration C(t) to work. Before building the
# full coupled system (Step 6-7), let's first understand Eq. 4 on its own by
# using a SIMPLE placeholder for C(t): just a constant drug level, always present.
# This isn't realistic yet (real drug levels rise and fall) - it's just to see
# how the drug term behaves before we add the more complex pharmacokinetics.

delta = 0.0005   # drug-kill rate constant (how strong the drug is per unit concentration)
C_constant = 20.0  # a constant, simplified drug concentration (placeholder for now)

def gompertz_with_drug_simple(t, N):
    N = max(N, 1e-8)
    growth = alpha * N * np.log(K / N)     # same normal growth as Eq. 1
    kill = delta * C_constant * N           # NEW: drug kills a portion of the tumor
    return growth - kill

# Solve it using the SAME RK4 function from before - no changes needed there
t_drug_simple, N_drug_simple = solve_rk4(gompertz_with_drug_simple, t0, tf, N0, dt)

print("Tumor volume at day 60:")
print("  Untreated (Standard Form):", N_standard[-1])
print("  Treated (constant drug level):", N_drug_simple[-1])
```

```python
# ---- Drug Pharmaco-kinetics (Eq. 5): dC/dt = -lambda * C(t) + u(t) ----
#
# This describes how drug concentration actually behaves in the body:
# it goes UP when a dose is given (u(t)), and goes DOWN steadily over
# time as the body eliminates it (-lambda * C).

lam = 0.3   # drug elimination rate (how fast the body clears the drug)
            # named "lam" instead of "lambda" because lambda is a reserved word in Python

# ---- The dosing function u(t): when and how much drug is given ----
dose_amount   = 5.0    # how much drug is delivered per dose
dose_interval = 7.0    # days between doses (e.g. once a week)
pulse_width   = 0.5    # how many days each dose "pulse" lasts (short = closer to a real injection)
first_dose_day = 5.0   # treatment starts on day 5

def u(t):
    if t < first_dose_day:
        return 0.0   # no drug given yet
    time_since_first = t - first_dose_day
    time_in_cycle = time_since_first % dose_interval   # where are we in the weekly cycle?
    if time_in_cycle < pulse_width:
        return dose_amount / pulse_width   # currently delivering a dose
    return 0.0   # between doses - no new drug being added

# ---- Eq. 5 written as code ----
def drug_pk(t, C):
    C = max(C, 0.0)   # drug concentration can't be negative
    return -lam * C + u(t)

# ---- Solve it on its own first, just to see what the concentration looks like ----
C0 = 0.0   # no drug in the body at the start
t_pk, C_pk = solve_rk4(drug_pk, t0, tf, C0, dt)

# Plot it
plt.figure(figsize=(9, 4))
plt.plot(t_pk, C_pk, color="green")
plt.axvline(first_dose_day, color="red", linestyle=":", label="Treatment starts")
plt.xlabel("Day")
plt.ylabel("Drug Concentration C(t)")
plt.title("Drug Concentration Over Time (Eq. 5 alone)")
plt.legend()
plt.show()
```

```python
# ---- Coupled System: tumor (Eq. 4) AND drug (Eq. 5) solved TOGETHER ----
#
# Until now, we solved N and C separately. But really, they affect each
# other: C(t) determines how hard the drug hits the tumor in Eq. 4,
# and both are changing at the same time. So now we track BOTH together
# as a pair, at every single time step.

def coupled_system(t, y):
    N, C = y   # unpack: y is now a PAIR of numbers [tumor volume, drug concentration]
    N = max(N, 1e-8)
    C = max(C, 0.0)

    dN = alpha * N * np.log(K / N) - delta * C * N   # Eq. 4 - now uses the REAL changing C(t)
    dC = -lam * C + u(t)                              # Eq. 5 - same as Step 6

    return np.array([dN, dC])   # return BOTH rates of change together

# ---- RK4 solver, adapted slightly to handle a PAIR of values instead of one ----
def solve_rk4_system(f, t0, tf, y0, dt):
    n_steps = int(np.round((tf - t0) / dt))
    t_vals = np.linspace(t0, tf, n_steps + 1)
    y_vals = np.zeros((n_steps + 1, len(y0)))   # a table: one row per time step, one column per variable (N and C)
    y_vals[0] = y0

    for i in range(n_steps):
        t_i, y_i = t_vals[i], y_vals[i]
        k1 = f(t_i, y_i)
        k2 = f(t_i + dt/2, y_i + dt/2 * k1)
        k3 = f(t_i + dt/2, y_i + dt/2 * k2)
        k4 = f(t_i + dt,   y_i + dt * k3)
        y_vals[i+1] = y_i + (dt/6) * (k1 + 2*k2 + 2*k3 + k4)

    return t_vals, y_vals

# ---- Solve the full coupled system ----
y0 = [N0, 0.0]   # starting point: tumor = N0, drug concentration = 0
t_coupled, y_coupled = solve_rk4_system(coupled_system, t0, tf, y0, dt)

N_coupled = y_coupled[:, 0]   # first column = tumor volume over time
C_coupled = y_coupled[:, 1]   # second column = drug concentration over time

print("Final tumor volume at day 60:")
print("  Untreated:", N_standard[-1])
print("  Treated (constant drug, Step 5):", N_drug_simple[-1])
print("  Treated (REAL changing drug, Step 7):", N_coupled[-1])
```

```python
"""
02_form_equivalence_and_rk4_check.py

Two separate validation figures:

(A) Standard Form (Eq. 1) vs Retardation Form (Eq. 2) vs Analytical (Eq. 3)
    -> confirms the beta = alpha/ln(K) reparameterization is algebraically
       identical to the original alpha-parameterized equation.
    Reproduces: Figure3_form_equivalence.png

(B) RK4 numerical solution plotted directly against the exact analytical
    solution (a more focused, two-curve version of the check in script 01).
    Reproduces: Figure2_RK4_validation.png
"""

import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------
# Parameters (identical to script 01 so results are directly comparable)
# ----------------------------------------------------------------------
alpha = 0.08
K     = 1500.0
N0    = 45.0
t_max = 60.0
dt    = 0.1
beta  = alpha / np.log(K)   # retardation parameter (Eq. 2)


def gompertz_analytical(t, N0, K, alpha):
    return K * np.exp(np.log(N0 / K) * np.exp(-alpha * t))


def rhs_standard(t, N, alpha, K):
    N = max(N, 1e-12)
    return alpha * N * np.log(K / N)


def rhs_retardation(t, N, alpha, beta):
    # dN/dt = N * [alpha - beta*ln(N)]
    N = max(N, 1e-12)
    return N * (alpha - beta * np.log(N))


def rk4_manual(f, t0, N0, t_max, dt, *args):
    n_steps = int(round((t_max - t0) / dt))
    t_vals = np.linspace(t0, t_max, n_steps + 1)
    N_vals = np.zeros(n_steps + 1)
    N_vals[0] = N0
    N = N0
    for i in range(n_steps):
        t = t_vals[i]
        k1 = f(t,          N,               *args)
        k2 = f(t + dt / 2, N + dt / 2 * k1,  *args)
        k3 = f(t + dt / 2, N + dt / 2 * k2,  *args)
        k4 = f(t + dt,     N + dt * k3,      *args)
        N = N + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
        N_vals[i + 1] = N
    return t_vals, N_vals


t_dense = np.linspace(0, t_max, 600)
N_analytical = gompertz_analytical(t_dense, N0, K, alpha)

t_std, N_std = rk4_manual(rhs_standard, 0, N0, t_max, dt, alpha, K)
t_ret, N_ret = rk4_manual(rhs_retardation, 0, N0, t_max, dt, alpha, beta)

# ----------------------------------------------------------------------
# Figure A: Standard vs Retardation vs Analytical
# ----------------------------------------------------------------------
plt.figure(figsize=(9, 6))
plt.plot(t_dense, N_analytical, color="gray", lw=6, alpha=0.6, label="Analytical (Eq. 3)")
plt.plot(t_std, N_std, color="blue", lw=2, ls="--", label="Eq. 1 (Standard Form)")
plt.plot(t_ret, N_ret, color="red", lw=2, ls=":", label=f"Eq. 2 (Retardation Form, beta={beta:.4f})")
plt.axhline(K, color="gray", ls=":", lw=1.2, label=f"K = {K}")
plt.xlabel("Day")
plt.ylabel("Tumor Volume N(t) (mm3)")
plt.title("Figure 3: Equivalence of Standard Form (Eq.1) and Retardation Form (Eq.2)")
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig("Figure3_form_equivalence.png", dpi=150)
print("Saved Figure3_form_equivalence.png")
plt.close()

# ----------------------------------------------------------------------
# Figure B: RK4 vs Exact Analytical Solution (focused two-curve check)
# ----------------------------------------------------------------------
plt.figure(figsize=(9, 6))
plt.plot(t_dense, N_analytical, color="gray", lw=6, alpha=0.6, label="Analytical (Eq. 3, exact)")
plt.plot(t_std, N_std, color="blue", lw=2, ls="--", label="RK4 numerical solution")
plt.axhline(K, color="gray", ls=":", lw=1.2, label=f"K = {K}")
plt.xlabel("Day")
plt.ylabel("Tumor Volume N(t) (mm3)")
plt.title("Figure 2: RK4 Solver Validated Against Exact Analytical Solution")
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig("Figure2_RK4_validation.png", dpi=150)
print("Saved Figure2_RK4_validation.png")

# Quick numerical agreement check
N_std_interp = np.interp(t_dense, t_std, N_std)
N_ret_interp = np.interp(t_dense, t_ret, N_ret)
print(f"Max abs diff (standard vs analytical):    {np.max(np.abs(N_std_interp - N_analytical)):.6f}")
print(f"Max abs diff (retardation vs analytical): {np.max(np.abs(N_ret_interp - N_analytical)):.6f}")
```
